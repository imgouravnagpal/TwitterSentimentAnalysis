```python
import numpy as np
import pandas as pd
Traindata = pd.read_csv('Train.csv')
Testdata = pd.read_csv('Test.csv')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
Traindata
```

```python
XTrain = [txt for txt in Traindata.text]
YTrain = [sentiment for sentiment in Traindata.airline_sentiment]
XTest = [txt for txt in Testdata.text]
```

```python
count_vec = CountVectorizer(max_features = 2000,max_df=0.8, min_df=0.001)
```

```python
X_Train_Transform = count_vec.fit_transform(XTrain)
X_Train_Transform
```

```python
count_vec.get_feature_names()
```

Count vectorizer will remove all the stop words along with puntuations .. no need to do it manually

```python
X_Test_Transform = count_vec.transform(XTest) 
```

```python
from sklearn.svm import SVC  #classifier
```

```python
classifier = SVC()
classifier.fit(X_Train_Transform,YTrain) #fitting into classifier
```

```python
YTest = classifier.predict(X_Test_Transform) 
```

```python
classifier.score(X_Test_Transform,YTest) #score is pretty good
```

```python
np.savetxt('predictions.csv',YTest,'%s')
```
